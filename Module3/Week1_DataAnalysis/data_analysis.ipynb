## B. Data Analysis with Time Series data


### B.1 Read dataset


```python
df = pd.read_csv(os.path.join(data_root_dir, 'opsd_germany_daily.csv'))
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
```

```python
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day_name()
```

```python
df.head()
```

### B.2 Time-based indexing


```python
df.loc['2017-08-15': '2017-08-20']
```

```python
df.loc['2017-08']
```

### B.3 Visualizing time series data


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize': (11, 4)})
df['Consumption'].plot(linewidth=0.5)
```

```python
col_list = ['Consumption', 'Solar', 'Wind']
axes = df[col_list].plot(marker='.', alpha=0.5,
                         linestyle='None', figsize=(11, 9), subplots=True)
for ax in axes:
    ax.set_ylabel('Daily Totals (GWh)')
plt.show()
```

### B.4 Seasonality


```python
fig, axes = plt. subplots(3, 1, figsize=(11, 10), sharex=True)
for name, ax in (zip(['Consumption', 'Solar', 'Wind'], axes)):
    sns.boxplot(data=df, x='Month', y=name, ax=ax)
    ax.set_ylabel('GWh')
    ax.set_title(name)
    if ax != axes[-1]:
        ax.set_xlabel('')
```

### B.5 Frequencies


```python
pd.date_range('2017-01-01', '2017-01-6', freq='D')
```

```python
time_samples = pd.to_datetime(['2013-03-03', '2013-03-06', '2013-03-08'])
consumption_samples = df.loc[time_samples, ['Consumption']].copy()
consumption_samples
```

```python
consumption_freq = consumption_samples.asfreq('D')
consumption_freq
```

```python
consumption_freq['Consumption - Forward Fill'] = consumption_samples.asfreq(
    'D', method='ffill')
consumption_freq
```

```python
consumption_freq['Consumption - Linaer'] = consumption_samples.asfreq(
    'D').interpolate(method='linear')
consumption_freq
```

### B.6 Resampling


```python
data_cols = ['Consumption', 'Solar', 'Wind']
df_weekly_mean = df[data_cols].resample('W').mean()
df_weekly_mean.head()
```

```python
start, end = '2017-01', '2017-06'
fig, axes = plt.subplots()
axes.plot(df.loc[start:end, 'Consumption'], label='Daily',
          marker='.', linestyle='-', linewidth=0.5)
axes.plot(df_weekly_mean.loc[start:end, 'Consumption'], marker='o',
          markersize=8, linestyle='-', linewidth=1, label='Weekly Mean Consumption')
axes.set_ylabel('Daily Totals (GWh)')
axes.legend()
plt.show()
```

```python
df_annual = df[data_cols].resample('A').sum(min_count=360)
df_annual.set_index(df_annual.index.year, inplace=True)
df_annual.index.name = 'Year'
df_annual.tail(5)
```

### B.7 Rolling windows


```python
df_7d = df[['Consumption']].rolling(window=7, center=True).mean()
df_7d.head(10)
```

### B.8 Trend


```python
df_365 = df[['Consumption']].rolling(
    window=365, center=True, min_periods=183).mean()
df_365.head(10)
```

```python
fig, axes = plt.subplots()
axes.plot(df['Consumption'], marker='.', markersize=2,
          color='black', linestyle='None', label='Daily')
axes.plot(df_7d, linewidth=2, label='7 days rolling mean')
axes.plot(df_365, linewidth=3, label='365 days rolling mean')
axes.set_xlabel('Year')
axes.set_ylabel('Daily Totals (GWh)')
axes.legend()
plt.show()
```
